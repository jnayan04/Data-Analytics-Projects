```python
# we will analyse data based on reservation_status_date,hence we will convert its data type to datetime which is initially an object type.

df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'],format="mixed") 
```

```python
# return column name and count of missing values in that column:-

df.isnull().sum()
```

```python
#since agent and company columns have so many missing values therefore we will remove those columns.Moreover these columns are not of significant use to the analysis that we are aiming for

df.drop(['company','agent'],axis=1,errors="ignore",inplace=True) # to remove agent and company columns
df.dropna(inplace=True)#to remove the missing values --to remove rows with missing values
```

```python
df.isnull().sum() # now the dataset has no missing values
```

```python
# summary statistics of the numerical columns: 

df.describe()
```

```python
# removing relevant outliers like adr being -6 or 5400

df=df[df['adr']>0]
df=df[df['adr']<5400]
```

```python
df.describe()
```

# Data Analysis and Visualisations:-

```python
# 'value_counts' function-->>returns category name and how many times it is present in the column:- use "(normalize=True)" in  value_counts() for percentage

cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc

#0--represents not cancelled
#1--represents cancelled
```

```python
# the above result implies that around 62% reservations are not cancelled and 37% of cancellations are cancelled
```

```python
# visualising the above result via a bar plot

cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))#figsize(width,height)
plt.title('Reservation Status Count')
plt.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
# finding cancellation rates depending on hotels

plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# above codes give cacnellation percentages of resort hotels and city hotels repectively
```

Effect of price on the cancellations:-

```python

resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# in whch month there are most cancellations or reservations:-

df['month']=df['reservation_status_date'].dt.month #extracting months from reservation status date
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
# cancellation: Highest=January;Lowest=August
#reservation: Highest=August;Lowest=January
```

```python

plt.figure(figsize=(15,8))
plt.title('ADR per Month',fontsize=30)
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())# month-wise avg daily rate ;'reset_index??--->>to take month column out from the index
plt.show()
```

```python
# the above plot shows that the adr is lowest in august which corresponds to lowest cancellations in the same month and similarly adr being highest in month of January,the cancellations are higher in january.
#Hence this proves that Higher prices lead to increased cancellations
```

```python
#country_wise

cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]#this command gives only the corresponding top 10 values filtered acc to the code
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)#autopct='%.2f' -->rounding off to decimal places
plt.show()
#figure shows portugal has highest no of cancellations
```

```python
# to check whose adr is high(cancelled reservations or not cancelled reservations)?

cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```
